### 04. Cleaning Data

##### 4.1 Missing Data

```python
# fill the NaN values with 'Type' - mode

df["Type"] = df['Type'].fillna(df['Type'].notnull().mode()[0])
```

```python
# fill the NaN values with 'Length' - avg

df["Length"] = df['Length'].fillna(df["Length"].notnull().mean())
```

```python
# fill the NaN values with 'Width' avg

df["Width"] = df['Width'].fillna(df["Width"].notnull().mean())
```

```python
# check the remaining null value

df.isnull().sum()
```

```python
# filter manufacturer, material and location missing values

df_filtered = df[df[["Manufacturer", "Material", "Location"]].notnull().all(axis=1)]
```

```python
# Check if there is any null values remaining in the new dataframe

df_filtered.isnull().sum()
```

```python
# check general information of the new filtered dataframe

df_filtered.info()
```

```python
df_filtered.columns
```

```python
# based on the info above, the index remain from 1 to 9887, so it's necessary to reset it

df_filtered.reset_index(drop=True, inplace=True)

df_filtered.info()
```

##### 4.2 Location Column

```python
# change encode

df_filtered.loc[:,'Location'] = df_filtered['Location'].astype(str).apply(lambda x: x.encode('latin1').decode('utf-8'))
```

```python
# check if the function was applied correctly

df_filtered.head()
```

```python
# considering that there is no pattern on how the location was defined, I will create another column with only the country

df_filtered.loc[:,'Country'] = df_filtered['Location'].str.split().str[0]
```

```python
# check if it was created and only contain countries

df_filtered['Country'].value_counts(dropna=False)
```

##### 4.3 - Price Column to US Dollars Currency

```python
# Considering how the price is shown, it is important to separate it first

df_filtered.loc[:,"Price_Currency"] = df_filtered["Price"].str.split().str[0].copy()
df_filtered.loc[:,"Amount"] = df_filtered["Price"].str.split().str[1].copy()
```

```python
# check 

df_filtered.head(10)
```

```python
# check dtypes

df_filtered.dtypes
```

```python
# change Price_Amount dtype

df_filtered.loc[:,'Amount'] = df_filtered['Amount'].astype(float)
```

```python
df_filtered["Price_Currency"].value_counts()
```

```python
# first define exchange rate for each currency in the column compared to EUR

rates = {"EUR": 1.10, "DKK": 1.10, "Â£": 1.30, "CHF": 0.15}

def rate_exchange(row):
  currency = row["Price_Currency"]
  amount = row["Amount"]
  
  if currency in rates:
    return amount * rates[currency]
  else: print("ValueError")

df_filtered["Amount_US"] = df_filtered.apply(rate_exchange, axis=1)
```

```python
df_filtered.head(10)
```

```python
df_filtered.columns
```

##### 4.4 - Year Built Column

```python
# Considering that the minimum year built it was 0, I will filter this dataframe to keep all the values that are !=0

df_filtered = df_filtered[df_filtered['Year_Built'] !=0]
```

```python
# check the amount of rows that remains after the cleaning

df_filtered.info()
```

##### 4.5 - Cleaning Dataframe After Changings

```python
# Information such as Location which don't follow a pattern, Price (including Currency + Amount), or just the Currency or Original Price 
# are not useful for our analysis, reason why I'll drop it

df_filtered.drop(columns=["Price", "Location", "Price_Currency", "Amount"], axis=1, inplace=True)
```

```python
df_filtered.reset_index(drop=True, inplace=True)
```
